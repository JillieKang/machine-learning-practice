# Supervised Machine Learning for Prediction

## Machine Learning Based on Linear Regression

```python
# Requirements
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Data set
mlnfb = pd.read_csv('c:/jilliepython/narcissism_and_fb_usage.csv')
mlnfb.head()
```

```python
# Defining feature and lable variables
X = mlnfb[['narcissism', 'age', 'gender']] # feature variable
y = mlnfb['fbuse'] # label variable
```

```python
# Spliting the dataset into training and test sets (30% test set, 70% training set)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Creating a linear regression model
model = LinearRegression()
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
# Checking coefficients
model.coef_
```

```python
# Making predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mse)
print(r2)

# R2: Moderate performance
# MSE: Nontrivial prediction error
```

## Ridge Regression

```python
# Regulararization to prevent overfitting
# Penalty for larger coefficients
# Reducing coefficient magnitude
```

```python
# Requirements

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Data set
df = pd.read_csv('c:/jilliepython/youtube_usage.csv')
```

```python
# Defining the features and target variable with the updated column names
X = df[['age', 'education', 'income', 'loneliness']]
y = df['youtubehrs']
```

```python
# Spliting the dataset into training and test sets (30% test set, 70% training set)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
# Standardizing the features (important for Ridge Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Ridge Regression with Lambda = 1

```python
# Creating and training the ridge regression model
ridge = Ridge(alpha=1) # define penalty term, aka lambda
ridge.fit(X_train_scaled, y_train)
```

```python
# Making predictions
y_pred_lambda1 = ridge.predict(X_test_scaled)
```

```python
# Checking coefficients from the ridge regression model
ridge_coefficients = ridge.coef_
ridge_coefficients
```

### Ridge Regression with Lambda = 100

```python
# Creating and training the ridge regression model
ridge = Ridge(alpha=100)
ridge.fit(X_train_scaled, y_train)
```

```python
# Making predictions
y_pred_lambda100 = ridge.predict(X_test_scaled)
```

```python
# Getting the coefficients from the Ridge model
ridge_coefficients = ridge.coef_
ridge_coefficients

# Increasing lambda reduced most coefficient magnitudes
```

```python
# Evaluating the model
mse_lambda100 = mean_squared_error(y_test, y_pred_lamda100)
r2_lambda100 = r2_score(y_test, y_pred_lamda100)
print(mse_lambda100)
print(r2_lambda100)

mse_lambda1 = mean_squared_error(y_test, y_pred_lambda1)
r2_lambda1 = r2_score(y_test, y_pred_lambda1)
print(mse_lambda1)
print(r2_lambda1)

# MSE decreased
# R2 increased
# Slightly improved generalization perforance
```
